```python
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(20,20))
for i, column in enumerate(data.columns, 1):
    plt.subplot(5,3,i)
    sns.distplot(data[column])
```

### Observation:-
* average number of rooms per dwelling is normally distributed(Skewed).
* pupil-teacher ratio by town and  proportion of owner-occupied units built prior to 1940 is left skewed.
* nitric oxides concentration, lower status of the population and weighted distances to five Boston employment centres are right skewed.

```python
plt.figure(figsize=(20,20))
for i, column in enumerate(data.columns, 1):
    plt.subplot(5,3,i)
    sns.scatterplot(x= data['MEDV'], y=data[column])
```

### Observations:-
* Mostly per capita crime rate by town is Less.
* proportion of residential land zoned for lots over 25,000 sq.ft. are mostly in 0.
* average number of rooms per dwelling have a positive relation with  Median value of owner-occupied homes

```python
# Checking for null Values

data.isnull().sum()
```

```python
for column in data.columns:
    data.fillna({
        column:np.mean(data[column])
    }, inplace=True)
```

```python
data.isnull().sum()
```

```python
# Checking is their any outliers using box plot

plt.figure(figsize=(20,8))
sns.boxplot(data=data, width=0.8);
```

### There are some outliers in data, so StandardScalar can help in scaling data.

```python
x = data.drop(columns='MEDV', axis=1)
y = data['MEDV']
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_scaled = scalar.fit_transform(x)
```

```python
X_scaled
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()

vif['VIF'] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]
vif['Features'] = x.columns

vif
```
